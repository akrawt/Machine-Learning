```python
import pandas as pd
import seaborn as sns
from sklearn import linear_model
```

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
lr = linear_model.LinearRegression()
```

```python
lr.fit(X = tips[["total_bill","size"]], y = tips["tip"])
```

```python
lr.coef_
# Interpretation:
# For every 1 dollar increase in total bill given the same size, the tip increases by coef[0] ( 0.0927...)
# For every 1 person increase in size given the same total_bill, the tip increases by coef[1] (0.19259779)
```

```python
lr.intercept_
```

```python
tips_dummy = pd.get_dummies(tips,drop_first = True)
```

```python
tips_dummy
```

```python
lr.fit(X = tips_dummy.iloc[:,2:], y = tips_dummy["tip"])
```
