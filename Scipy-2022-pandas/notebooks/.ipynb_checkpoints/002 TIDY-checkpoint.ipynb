```python
billboard = pd.read_csv("../data/billboard.csv")
```

```python
billboard.head()
```

```python
billboard.melt(id_vars = ["year","artist","track","time","date.entered"],
              var_name = "week",
              value_name = "rank")
```

```python
(
billboard
 .melt(id_vars = ["year","artist","track","time","date.entered"],
        var_name = "week",
        value_name = "rank")
 .groupby("artist")["rank"]
 .mean()
)
```

```python
# Variables are stored in rows and columns
```

```python
weather = pd.read_csv("../data/weather.csv")
```

```python
weather.head()
```

```python
weather_long = weather.melt(id_vars = ["id","year","month","element"],
            var_name = "day",
            value_name = "temp")
```

```python
weather_long.head()
```

```python
#Opposite of melt()
```

```python
weather_long.pivot_table(index = ["id","year","month","day"],
                        columns="element",
                        values="temp")
```

```python
(
weather_long.pivot_table(index = ["id","year","month","day"],
                        columns="element",
                        values="temp")
.reset_index()
)
```
